## Generate BLER vs EsNo table using PUSCH Matlab Rx

```python
np.random.seed(1234) # Set seed so results are consistent run to run
NUM_TBS = 10000
MIN_TB_ERRORS = 50
EsNo_dB_range = np.arange(3.0,6.54,0.5)
tb_count = []
tb_err_count = []

gNB = puschTxRx.matlab_rx_init()
python_txrx_helper.print_status_header()

for EsNo_idx, EsNo_dB in enumerate(EsNo_dB_range):
    sigma = np.sqrt(0.5*pow(10.,-EsNo_dB/10.))
    
    t_start = datetime.datetime.now()
    
    cur_tb_err_count = 0
    for cur_tb_count in range(NUM_TBS):
        rx_tensor = python_txrx_helper.awgn_channel(tx_tensor,sigma)
        matlabRxSamp = python_txrx_helper.matlab_rx_helper(rx_tensor, UE_out)
        
        gNB = puschTxRx.matlab_rx(matlabRxSamp)
        if (gNB['FAPIpdu'][0]['tbErr'] != 0.0):
            cur_tb_err_count = cur_tb_err_count + 1
            
        t_end = datetime.datetime.now()
        python_txrx_helper.print_status(EsNo_dB, cur_tb_count+1, cur_tb_err_count, t_end-t_start, False)
        if (cur_tb_err_count >= MIN_TB_ERRORS):
            break # out of cur_tb_count for loop
            
    python_txrx_helper.print_status(EsNo_dB, cur_tb_count+1, cur_tb_err_count, t_end-t_start, True)
    tb_count.append(cur_tb_count+1)
    tb_err_count.append(cur_tb_err_count)
```

## Generate BLER vs EsNo table using PUSCH cuPHY Rx

```python
np.random.seed(1234) # Set seed so results are consistent run to run
NUM_TBS = 10000
MIN_TB_ERRORS = 200
EsNo_dB_range = np.arange(3.0,6.54,0.5)
tb_count = []
tb_err_count = []

python_txrx_helper.print_status_header()

for EsNo_idx, EsNo_dB in enumerate(EsNo_dB_range):
    sigma = np.sqrt(0.5*pow(10.,-EsNo_dB/10.))

    t_start = datetime.datetime.now()
    
    # setup a CUDA stream
    stream = python_txrx_helper.check_cuda_errors(cudart.cudaStreamCreate())
    
    cur_tb_err_count = 0
    for cur_tb_count in range(NUM_TBS):
        rx_tensor = python_txrx_helper.awgn_channel(tx_tensor,sigma)
#         rx_tensor = np.real(rx_tensor).astype(np.half) + 1j*np.imag(rx_tensor).astype(np.half)
        tbOutput,crc = python_txrx_helper.test_PUSCH_Rx(rx_tensor, stream)
        if (crc != 0):
            cur_tb_err_count = cur_tb_err_count + 1
            
        t_end = datetime.datetime.now()
        python_txrx_helper.print_status(EsNo_dB, cur_tb_count+1, cur_tb_err_count, t_end-t_start, False)
        if (cur_tb_err_count >= MIN_TB_ERRORS):
            break # out of cur_tb_count for loop
            
    python_txrx_helper.print_status(EsNo_dB, cur_tb_count+1, cur_tb_err_count, t_end-t_start, True)
    tb_count.append(cur_tb_count+1)
    tb_err_count.append(cur_tb_err_count)
```
